```python

        
import random
from PIL import Image, ImageDraw, ImageFilter

def generate_random_pastel_gradient_image(img_size=(1000, 1000), border_thickness=20, output_path="border.png"):
    # Define a list of pastel tone colors
    pastel_colors = [
        (230, 200, 210),  # Soft Pink
        (200, 220, 230),  # Soft Blue
        (180, 230, 180),  # Soft Green
        (230, 210, 190),  # Soft Bisque
        (210, 180, 210),  # Soft Plum
        (220, 210, 220),  # Soft Thistle
        (240, 230, 240),  # Soft Lavender
        (230, 230, 240)   # Soft Alice Blue
    ]

    # Randomly select four colors for the gradient
    colors = random.sample(pastel_colors, 4)

    # Create a new image with RGBA mode
    image = Image.new("RGBA", img_size)

    # Create vertical gradient effect
    for y in range(img_size[1]):
        ratio = y / img_size[1]
        r = int((1 - ratio) * colors[0][0] + ratio * colors[1][0])
        g = int((1 - ratio) * colors[0][1] + ratio * colors[1][1])
        b = int((1 - ratio) * colors[0][2] + ratio * colors[1][2])
        ImageDraw.Draw(image).line([(0, y), (img_size[0], y)], fill=(r, g, b))

    # Add horizontal gradient layer
    for x in range(img_size[0]):
        ratio = x / img_size[0]
        r = int((1 - ratio) * colors[2][0] + ratio * colors[3][0])
        g = int((1 - ratio) * colors[2][1] + ratio * colors[3][1])
        b = int((1 - ratio) * colors[2][2] + ratio * colors[3][2])
        ImageDraw.Draw(image).line([(x, 0), (x, img_size[1])], fill=(r, g, b), width=1)

    # Apply Gaussian blur to smooth the gradients
    image = image.filter(ImageFilter.GaussianBlur(15))

    # Draw a white border
    border_color = (255, 255, 255, 0)
    draw = ImageDraw.Draw(image)
    draw.rectangle(
        [border_thickness, border_thickness, img_size[0] - border_thickness, img_size[1] - border_thickness],
        outline=border_color,
        width=border_thickness
    )

    # Draw a transparent fill space
    draw = ImageDraw.Draw(image)
    draw.rectangle(
        [border_thickness * 2.5, border_thickness * 2.5, img_size[0] - border_thickness * 2.5, img_size[1] - border_thickness * 2.5],
        fill=(255, 255, 255, 0)
    )

    # Save the image
    image.save(output_path)
    return output_path

# Generate and save the random pastel gradient image
file_path = generate_random_pastel_gradient_image()
file_path
```

```python
import time

# 랜덤 딜레이 함수
def _random_delay(self, min_delay=1, max_delay=3):
    time.sleep(random.uniform(min_delay, max_delay))

# 휴먼 액션: 마우스 이동
def _human_mouse_movement(self, element):
    action = ActionChains(self.driver)
    action.move_to_element(element).perform()
    self._random_delay(0.5, 1.5)

# 휴먼 액션: 페이지 스크롤
def _human_scroll(self, direction="down"):
    if direction == "down":
        self.driver.execute_script("window.scrollBy(0, 300);")
    elif direction == "up":
        self.driver.execute_script("window.scrollBy(0, -300);")
    self._random_delay(1, 2)
```
